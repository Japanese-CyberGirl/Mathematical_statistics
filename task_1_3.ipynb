### Написать генератор неоднородного процесса Пуассона $N(t)$ с заданной интенсивностью $\lambda(t)$ на интервале времени $[T_1, T_2]$.
### Ожидаемый результат - график траектории процесса $N(t)$.
### Провести серию из n (= 10, 100, 1000, 10000, ...) симуляций, по ним эмпирически оценить интесивность и сравнить её с заданной функцией интенсивности.

#### $\lambda(t) = \frac{1}{2+t}, t \in [0,1]$

Дадим сначала определение неоднородного процесса Пуассона $N(\cdot)$ с интенсивностью $\lambda(t)$. 

Этот процесс задается двумя свойствами:

1) Для функции $N(a,b) = N(b) - N(a)$, для любых $t, h > 0$ верно:   

$\mathbb{P}(N(t, t+h) = 0) = 1 - \lambda h + o(h), \ h \rightarrow 0$  
$\mathbb{P}(N(t, t+h) = 1) = \lambda h + o(h), \ h \rightarrow 0$  
$\mathbb{P}(N(t, t+h) \geq 2) = o(h), \ h \rightarrow 0$  

2) Количество точек $N(I_1),...,N(I_k)$ внутри **непересекающихся** конечных отрезков $I_1,...I_k$ являются **независимыми** случайными величинами.

Первое условие по существу требует, чтобы внутри очень маленького интервала длины $h$ находилась либо одна точка с вероятностью $\lambda h$, либо ни одной точки с вероятностью $1 - \lambda h$

Пусть $T_1$ - момент первого события. Тогда его функция распределения равна:   
$$F(t) = P(T_1 \leq t) = 1 - \exp(-\int_0^t \lambda(s) ds)$$

Тогда количеством событий неоднородного процесса Пуассона $N(t)$ на отрезке $[0, T]$ будет являться интеграл:  
$$\int_0^T \lambda(s)ds = \Lambda(t)$$

Положим, что функция $\Lambda(t)$ строго возрастает $\Longrightarrow \exists$ обратная функция $\Lambda^{-1}(t)$

Рассмотрим обратный процесс $M(s) = N(\Lambda^{-1}(s)), s \geq 0$

Покажем, что процесс $M(s)$ - является однородным Пуассоновским процессом с интенсивностью 1.

Возьмем произвольные $0 \leq s_1 \leq s_2$. Тогда  
$$M(s_2) - M(s_1) = N(\Lambda^{-1}(s_2)) - N(\Lambda^{-1}(s_1))$$


Так как $N$ - неоднородный Пуассоновский процесс, то приращение на интервале $(\Lambda^{-1}(s_1), \Lambda^{-1}(s_2))$  
Имеет распределение Пуассона с параметром $$\int_{\Lambda^{-1}(s_1)}^{\Lambda^{-1}(s_2)} \lambda(u) du$$

Следовательно $M(s_2) - M(s_1) \sim П(\int_{\Lambda^{-1}(s_1)}^{\Lambda^{-1}(s_2)} \lambda(u) du)$

По определнию функции $\Lambda$:
$$\int_{\Lambda^{-1}(s_1)}^{\Lambda^{-1}(s_2)} \lambda(u) du = \Lambda (\Lambda^{-1}(s_2) - \Lambda^{-1}(s_1))$$

И так как $\Lambda^{-1}$ - обратная к $\Lambda$, то

$\Lambda(\Lambda^{-1}(s_2)) = s_2$ , $\Lambda(\Lambda^{-1}(s_1)) = s_1$

Следовательно $$\int_{\Lambda^{-1}(s_1)}^{\Lambda^{-1}(s_2)} \lambda(u) du = s_2 - s_1$$

Таким образом $M(s_2) - M(s_1) \sim П(s_2 - s_1)$, то есть распределение приращения $M$ зависит только от длины интервала $s_2 - s_1$, что верно для однородного пуассоновского процесса с интенсивностью 1

Проверим независимость: рассмотрим непересекающиеся интервалы  
$(s_1, s_2], (s_3, s_4],...,(s_{2k-1},s_{2k}], \ s_1 < s_2 \leq s_3 < s_4 \leq s_5 < ... \leq s_{2k-1} < s_{2k}$

И так как функция $\Lambda^{-1}$ возрастает, следовательно, соответствующие интервалы: $(\Lambda^{-1}(s_i), \Lambda^{-1})(s_{i+1})]$ - не пересекаются.

Так как у процесса $N$ - независимые приращения, следовательно и случайные величины $N(\Lambda^{-1}(s_2)) - N(\Lambda^{-1}(s_1))$ - тоже независимы. И так как они в точности равны приращениям процесса $M$: $M(s_2) - M(s_1),...$ - то и у процесса $M$ - независимые приращения.

Таким образом для любых $0 \leq s_1 < s_2$: $M(s_2) - M(s_1) \sim П(s_2 - s_1)$  
И приращения процесса $M$ на непересеакющихся интервалах независимы.  

Следовательно, $M(s)$ - однородный процесс пуассона интенсивности 1

Тогда, полагая, $N(t) = M(\Lambda(t))$ получаем неоднородный процесс пуассона с интенсивностью $\lambda(t)$

Получается нам для построения неоднородного процесса Пуассона с интенсивностью $\lambda(t)$ нужно:
1) построить однородный процесс Пуассона $M(s)$ на интервале $[0, \Lambda(1)]$ с интенсивностью 1 (т.к. $t \in [0,1]$).
2) преобразовать его в неоднородный.

В нашей задаче $\lambda(t) = \frac{1}{2 + t}$, $t \in [0, 1]$

Тогда $$\Lambda(t) = \int_0^t\frac{1}{2+u}du = \ln(2+t) - \ln2 = \ln \frac{2+t}{2}$$

Функция $\Lambda(t)$ строго возрастает на $[0,1]$, найдем обратную к ней. Пусть $s = \Lambda(t) = \ln\frac{2+t}{2}$

Тогда $e^s = \frac{2+t}{2} \Longrightarrow 2e^s = 2+t, t=2e^s-2=2(e^s-1) \Longrightarrow \Lambda^{-1}(s)=2(e^s-1)$

Заметим, что $\Lambda(1) = \ln\frac{3}{2} \Longrightarrow$ задаем однородный пуассоновский процесс интенсивности 1 на интервале $[0, \ln\frac{3}{2}]$

Моменты перескоков однородного пуассоновского процесса - это суммы независимых экспоненциальных случайных величин:  
$E_1, E_2, ... , \sim Exp(1) \Longrightarrow S_1 = E_1, S_2 = E_1 + E_2,...$

Нас интересуют только такие значения $S_k$, что $S_k < \Lambda(1)$, в результате мы получаем набор моментов $T_k = \Lambda^{-1}(S_k) \Longrightarrow$  
$\Longrightarrow$ число событий к моменту $t$ равно $$N(t) = \sum_{k \geq 1} \mathbb{I}_{T_k\leq t}$$

Перейдем к написанию кода.

```python
import numpy as np
import matplotlib.pyplot as plt

def lam(t): #интенсивность неоднородного процесса
    return 1 / (2 + t)

def Lambda(t): #интенсивность однородного процесса
    return np.log((2 + t) / 2)

def Lambda_inv(s): #обратное преобразование времени
    return 2 * (np.exp(s) - 1)

T = 1.0
L = Lambda(T)

# моменты перескока для однородного процесса на [0, L]
def simulate_hpp_jumps(L):
    jumps = []
    s = 0.0
    while True:
        s += np.random.exponential(scale=1.0)
        if s >= L:
            break
        jumps.append(s)
    return np.array(jumps)

# моменты перескока для неоднородного процесса на [0, 1]
def simulate_nhpp_jumps():
    s_jumps = simulate_hpp_jumps(L)
    t_jumps = Lambda_inv(s_jumps)
    return t_jumps
```

```python
# одна реализация процесса
jumps = simulate_nhpp_jumps()

t_grid = np.linspace(0, T, 1000)

# значение N(t)
N_t = np.searchsorted(jumps, t_grid, side="right")

plt.figure(figsize=(8, 4))
plt.step(t_grid, N_t, where="post")
plt.xlabel("t")
plt.ylabel("N(t)")
plt.title("Одна траектория неоднородного пуассоновского процесса")
plt.grid(True)
plt.show()

print("Моменты перескоков:", jumps)
print("Число событий на [0,1]:", len(jumps))
```

Произведем оценку интенсивности: разобьем искомый интервал $[0,1]$ на отрезки $\Delta t$, для каждого из них подсчитаем количество событий, попавших в него, а затем усредним по количеству симуляций.

Таким образом получим для каждого промежутка его интенсивность: $\lambda_{new} = \frac{1}{n \Delta t} \sum^{n}_{i=1}N_i{\mathbb{I_j}}$ , где $N_i(\mathbb{I}_j) - число событий на $j-м$ интервале в $i-й$ симуляции.

За теоретическую же интенсивность обозначим искомую $\lambda(t) = \frac{1}{2+t}$

```python
def estimate_intensity(n_sim=1000, n_bins=40):
    bins = np.linspace(0, T, n_bins + 1)
    counts = np.zeros(n_bins)

    for _ in range(n_sim):
        jumps = simulate_nhpp_jumps()
        hist, _ = np.histogram(jumps, bins=bins)
        counts += hist

    dt = bins[1] - bins[0]
    centers = 0.5 * (bins[:-1] + bins[1:])
    lambda_new = counts / (n_sim * dt)

    return centers, lambda_new
```

```python
x = np.linspace(0, 1, 400)

for n in [10, 100, 1000, 10000]:
    centers, lambda_hat = estimate_intensity(n_sim=n, n_bins=40)

    plt.figure(figsize=(8, 4))
    plt.plot(x, lam(x), label="теоретическая λ(t)")
    plt.plot(centers, lambda_hat, marker="o", linestyle="-", label=f"эмпирическая оценка, n={n}")
    plt.xlabel("t")
    plt.ylabel("интенсивность")
    plt.title(f"Сравнение теоретической и эмпирической интенсивности, n={n}")
    plt.grid(True)
    plt.legend()
    plt.show()
```

Теперь построим график кумулятивной суммы событий, а не интенсивности процесса.

```python
def simulate_N_trajectory():
    jumps = simulate_nhpp_jumps()
    t_grid = np.linspace(0, T, 200)
    N_t = np.searchsorted(jumps, t_grid, side="right")
    return t_grid, N_t

# аналогично с вычислением интенсивности накапливаем на каждом интервале значения, после чего нормируем по количеству симуляций
def empirical_mean_N(n_sim):
    t_grid = np.linspace(0, T, 200)
    N_sum = np.zeros_like(t_grid)

    for _ in range(n_sim):
        _, N_t = simulate_N_trajectory()
        N_sum += N_t

    return t_grid, N_sum / n_sim


for n in [2, 3, 5, 10, 100, 1000, 10000]:

    t_grid, mean_N = empirical_mean_N(n)
    Lambda_vals = Lambda(t_grid)

    plt.figure(figsize=(8,4))
    plt.plot(t_grid, Lambda_vals, color="red", label="Теоретическая Λ(t)")
    plt.plot(t_grid, mean_N, color="blue", label="Эмпирическое среднее N(t)")
    plt.title(f"{n} симуляций")
    plt.xlabel("t")
    plt.ylabel("значение")
    plt.legend()
    plt.grid(True)
    plt.show()
```
